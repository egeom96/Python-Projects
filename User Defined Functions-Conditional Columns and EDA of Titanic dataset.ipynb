```python
female_avg_tip=tips[tips["sex"]=="Female"]["total_bill"].mean() #FEMALE TOTAL_BILL AVERAGE
```

```python
male_avg_tip=tips[tips["sex"]=="Male"]["total_bill"].mean() #MALE TOTAL_BILL AVERAGE
```

```python
male_avg_tip
```

```python
female_avg_tip
```

```python
conditions = [
    (tips['sex'] == 'Female') & (tips['total_bill'] < female_avg_tip),
    (tips['sex'] == 'Female') & (tips['total_bill'] >= female_avg_tip),
    (tips['sex'] == 'Male') & (tips['total_bill'] < male_avg_tip),
    (tips['sex'] == 'Male') & (tips['total_bill'] >= male_avg_tip)
]

values=[0,1,0,1]

tips["total_bill_flag"]=np.select(conditions,values)
```

```python
def conversion(df,x):
    df[x,"in euros"]=df[x]*0.85
    k= df[x,"in euros"]
    df.replace("tip",k)
    return df
```

```python
conversion(tips1,"tip")
```

```python
def total_bill_segments(df,x,y):
    for index,row in df.iterrows():
        if row[x]<15:
            df.loc[index,y]="Low Bill"
        elif row[x]<20:
              df.loc[index,y]="Medium Bill"
        elif row[x]>=20:
             df.loc[index,y]="High Bill"
    return df
            
```

```python
total_bill_segments(tips1,"total_bill","bill_segment")
```

```python
def avg_size(df):
    for index,row in df.iterrows():
        if row["size"]<=2:
            df.loc[index,"Size Category"]="Small"
            df.loc[index,"Tip Percentage"]=row["tip"]/row["total_bill"]*100
        else:
            df.loc[index,"Size Category"]="Large"
            df.loc[index,"Tip Percentage"]=row["tip"]/row["total_bill"]*100
    print(df)
    return df.groupby("Size Category")["Tip Percentage"].mean()
```

```python
avg_size(tips1)
```

```python
def tip_by_gender(df):
    avg_male=df[df["sex"]=="Male"]["tip"].mean()
    avg_female=df[df["sex"]=="Female"]["tip"].mean()
    overall=df["tip"].mean()
    
    if avg_male<overall:
        print("avg male tip is below avg overall tip")
    elif avg_male>overall:
        print("avg male tip is above avg overall tip")
    
    if avg_female<overall:
        print("avg female tip is below avg overall tip")
    elif avg_female>overall:
        print("avg female tip is above avg overall tip")
```
