```python
#Q10: Count the number of countries for which the total_tests data is missing.
a=covid_df[covid_df["total_tests"].isna()]
a[["location"]].count()
```

```python
#Q11: Merge countries_df with covid_data_df on the location column.
merged_df=covid_df.merge(countries_df,on="location")
merged_df
```

```python
#Q12: Add columns tests_per_million, cases_per_million and deaths_per_million into combined_df.

for index,row in merged_df.iterrows():
    merged_df.at[index,"tests_per_million"]=row["total_tests"]/row["population"]*1000000
    merged_df.at[index,"cases_per_million"]=row["total_cases"]/row["population"]*1000000
    merged_df.at[index,"deaths_per_million"]=row["total_deaths"]/row["population"]*1000000
```

```python
#13-14-15 Create a dataframe with 10 countires that have highest number of tests,deaths and cases per million people.
a =merged_df.sort_values("tests_per_million",ascending=False).head(10)
a
```

```python
b=merged_df.sort_values("cases_per_million",ascending=False).head(10)
b
```

```python
c=merged_df.sort_values("deaths_per_million",ascending=False).head(10)
c
```

```python
#(Optional) Q: Count number of countries that feature in both the lists of "highest number of tests per million" and "highest number of cases per million".

count=0

for index_a,row_a in a.iterrows():
    for index_b,row_b in b.iterrows():
        if a.at[index_a,"location"]==b.at[index_b,"location"]:
            count=count+1
        else:
            count
            
print(count)
```
