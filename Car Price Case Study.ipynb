#3b) The threshold correlation value is selected as 0.75, which variables will you include in the model? Are there any variable pairs that might be affected by multicollinearity? If so, how will you deal with this issue?

When we talk about multicollinearity, we're primarily concerned with the relationships among predictor variables (features), 
not necessarily between a predictor variable and the target variable.

```python
#3b cont. found multicollinear independent variables and their importance with the use of VIF method that indicates strong rship >5
# features to consider removing (greater than 0.75 treshold)
considered_features = ["carlength","carwidth","curbweight","enginesize","citympg","highwaympg","horsepower","wheelbase"]

# load statmodels functions
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# compute the vif for all given features
def compute_vif(considered_features):
    
    X = carprice_df[considered_features]
    # the calculation of variance inflation requires a constant
    X['intercept'] = 1
    
    # create dataframe to store vif values
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif
compute_vif(considered_features).sort_values('VIF', ascending=False)
```

```python
considered_features.remove('citympg')
compute_vif(considered_features)
```

```python
considered_features.remove('curbweight')
compute_vif(considered_features)
```

```python
considered_features.remove('carlength')
compute_vif(considered_features)
```

```python
considered_features.remove('horsepower')
compute_vif(considered_features)
```

```python
#3-c Now that you are done with feature selection, prepare the features for the model, (Hint: you need to “scale” your variables) and split the dataset into test-train sets (30% and 70% respectively)

#As per company policy categorical variables not taken into consideration.So I am moving only with remaining numerical variables
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assume X contains the independent variables and y contains the dependent variable
X = carprice_df.loc[:, ['wheelbase', 'carheight','enginesize','boreratio', 'stroke', 'compressionratio',
        'peakrpm', 'highwaympg']]

y = carprice_df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#3-d Perform multiple linear regression? What are the model results? Are there any variables with p-values larger than 0.05? If so, how should you proceed? Please comment on your findings

import statsmodels.api as sm

# Add a constant to the X_train_scaled and X_test_scaled datasets
X_train_scaled_const = sm.add_constant(X_train_scaled)
X_test_scaled_const = sm.add_constant(X_test_scaled)

# Fit the ordinary least squares (OLS) model
model = sm.OLS(y_train, X_train_scaled_const).fit()

# Print the model summary
print(model.summary())
```

Interpretation:
R-squared & Adjusted R-squared: The R-squared value is 0.847, indicating that approximately 84.7% of the variability in the dependent variable (price) is explained by the independent variables in the model. The Adjusted R-squared is 0.838, which adjusts for the number of predictors in the model, slightly penalizing for the inclusion of non-significant predictors.

F-statistic & Prob (F-statistic): The F-statistic is 92.61 with a very low p-value (Prob (F-statistic) = 8.52e-51), suggesting that the model as a whole is statistically significant.

Coefficients:

The intercept (const) is statistically significant with a p-value of 0.000.
The coefficients for x1, x2, x3, x4, x5, x6, x7, and x8 represent the effect of each variable on the dependent variable, 'price'.
Among these, only the coefficients for x3, x5, x6, x7, and x8 are statistically significant (p-values < 0.05). This implies that these variables have a statistically significant relationship with 'price' at the 5% significance level.
Variables with p-values greater than 0.05 (e.g., x1, x2, and x4) are not statistically significant predictors of 'price' at the 5% significance level.

```python
#4-e To automize the model process for later use, can you create a model pipeline that includes the following steps:
#To automize the model process for later use, can you create a model pipeline that includes the following steps:
#a. A class that performs all the necessary feature engineering & selection
#b. Scales the data accordingly
#c. Gets model output
#How would you fit the training dataset and then get a prediction for test dataset using this pipeline?
```

```python
#4e-a This class will handle any feature engineering tasks to perform, such as dropping unnecessary columns, handling missing values, etc.

from sklearn.base import BaseEstimator, TransformerMixin

class FeatureEngineering(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        # Check if columns exist before dropping them
        columns_to_drop = ['car_ID', 'CarName']
        existing_columns = [col for col in columns_to_drop if col in X.columns]
        
        # Drop columns if they exist
        if existing_columns:
            X = X.drop(columns=existing_columns)
        
        return X
```

```python
#4e-b to define a pipeline that includes feature engineering, scaling, and the linear regression model.

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# Initialize the pipeline with steps
pipeline = Pipeline([
    ('feature_engineering', FeatureEngineering()),  # Step 1
    ('scaler', StandardScaler()),  # Step 2: Scaling
    ('model', LinearRegression())  # Step 3: Model
])
```

```python
#4e-b Fit the pipeline to the training data

pipeline.fit(X_train, y_train)
```

```python
# 4e-c Predict on the test data using the pipeline.Once the pipeline is fitted, it can be used it to make predictions on the test dataset.
y_pred = pipeline.predict(X_test)
```
